```python
import pandas as pd
import random
```

```python
# Define categories and subcategories
categories = {
    "Workplace Culture": ["Team Collaboration", "Leadership", "Inclusivity"],
    "Work-Life Balance": ["Flexibility", "Remote Work", "Time Off"],
    "Career Growth": ["Promotions", "Training Programs", "Mentorship"],
    "Compensation": ["Salary", "Bonuses", "Benefits"],
    "Work Environment": ["Facilities", "Office Setup", "Ergonomics"],
    "Management": ["Communication", "Support", "Decision Making"],
    "Job Role": ["Responsibilities", "Expectations", "Clarity"]
}

# Example open responses
response_templates = {
    "Team Collaboration": "I enjoy working closely with my team on projects.",
    "Leadership": "Our leaders are approachable and provide clear direction.",
    "Inclusivity": "The company promotes an inclusive culture for everyone.",
    "Flexibility": "I appreciate the flexible working hours offered.",
    "Remote Work": "Working remotely has been a seamless experience.",
    "Time Off": "The leave policies are generous and easy to access.",
    "Promotions": "There is a clear path for promotions and recognition.",
    "Training Programs": "The company invests in training and skill development.",
    "Mentorship": "Having a mentor has helped me grow professionally.",
    "Salary": "The salary is competitive with industry standards.",
    "Bonuses": "Bonuses are performance-based and fairly distributed.",
    "Benefits": "We receive comprehensive health and wellness benefits.",
    "Facilities": "Office facilities are well-maintained and resourceful.",
    "Office Setup": "The workspace setup supports productivity and comfort.",
    "Ergonomics": "Ergonomic furniture helps with long working hours.",
    "Communication": "Management communicates updates regularly.",
    "Support": "I feel supported by my managers in my daily tasks.",
    "Decision Making": "Leaders involve teams in the decision-making process.",
    "Responsibilities": "My job responsibilities are clear and well-defined.",
    "Expectations": "Expectations from my role are communicated effectively.",
    "Clarity": "There is clarity in what is expected from my position."
}
```

```python
# Generate 100 samples
data = []
for _ in range(100):
    category = random.choice(list(categories.keys()))
    subcategory = random.choice(categories[category])
    response = response_templates.get(subcategory, f"Feedback about {subcategory}.")
    data.append({
        "answer": response,
        "category": category,
        "subcategory": subcategory
    })

df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("input_train_data.csv", index = False)
```

```python
# Generate new answers using the same subcategories for a realistic test set
new_data = []
for _ in range(100):
    subcategory = random.choice(list(response_templates.keys()))
    response = response_templates[subcategory] + " " + random.choice([
        "This has made a big difference in my day-to-day work.",
        "I believe this is an area where the company excels.",
        "It's been consistently good since I joined.",
        "This stands out compared to my previous job.",
        "It contributes positively to my overall experience."
    ])
    new_data.append({
        "answer": response
    })

new_df = pd.DataFrame(new_data)
```

```python
new_df.to_csv("input_new_data.csv", index = False)
```
